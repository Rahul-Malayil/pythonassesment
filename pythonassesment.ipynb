```python
import pandas as pd
import numpy as np
import re
import array as ar
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/project.csv.csv")
print(data)
```

```python
team_count = data["Name"].value_counts()
percentage = (team_count / len(data)) * 100

print("Team Distribution:")
print(team_count)
print("Team Percentages:")
print(percentage)



sns.countplot(x="Name", data=data)
plt.title("Team Distribution")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#2
position = data["Position"].value_counts()

print("Position Distribution:")
print(position)



sns.countplot(x="Position", data=data)
plt.title("Position Distribution")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#3
ages = data["Age"].value_counts()

print("Age Group Distribution:")
print(ages)



sns.histplot(x="Age", data=data)
plt.title("Age Group Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#4
highestsalary = data.loc[data["Salary"].idxmax()]

print("Highest Salary Expenditure:")
print(highestsalary)



sns.barplot(x="Salary", y="Name", data=data.nlargest(5, "Salary"))
plt.title("Top 5 Salaries")
plt.xlabel("Salary")
plt.ylabel("Name")
plt.show()
```

```python
#5
correlation = data["Age"].corr(data["Salary"])

print("Correlation Coefficient:")
print(correlation)



sns.scatterplot(x="Age", y="Salary", data=data)
plt.title("Age & Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
